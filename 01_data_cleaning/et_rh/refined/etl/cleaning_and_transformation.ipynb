```python
modified_df = df.withColumn(
    "Education Level",
    F.when(F.col("Education Level") == "Master’s Degree", "Masters Degree")
    .when(F.col("Education Level") == "Bachelor’s Degree", "Bachelors Degree")
    .otherwise(F.col("Education Level"))
)
```

```python
# Leitura do dicionário
dicionario_path = "/home/tatiane/data-projects/01_data_cleaning/et_rh/refined/metadata/dicionario.csv"
dic = spark.read.csv(dicionario_path, header=True, inferSchema=False)
```

```python
dic.show()
```

```python
import json
from pyspark.sql.functions import col, udf

def safe_json_loads(s):
    if not s or s.strip() in ["", "{}", "null", None]:
        return {}
    try:
        return json.loads(s)
    except json.JSONDecodeError:
        # Corrige strings com aspas simples
        try:
            return json.loads(s.replace("'", '"'))
        except Exception:
            return {}
```

```python
from pyspark.sql.types import ByteType, IntegerType, StringType

dict_rows = dic.collect()
for row in dict_rows:
    old_name = row["var"]
    new_name = row["new_var"]
    tipo = str(row["new_type"]).lower() if row["new_type"] else None
    mapping = safe_json_loads(row["map_values"])

    # Renomear coluna
    if old_name in modified_df.columns:
        modified_df = modified_df.withColumnRenamed(old_name, new_name)

    if mapping:
        # Define função para substituir valores
        #print(f"Mapeando {new_name}: {mapping}")
        mapping_udf = udf(lambda x: mapping.get(str(x), x), StringType())
        modified_df = modified_df.withColumn(new_name, mapping_udf(col(new_name)))
```

```python
## Mudando o tipo das variáveis
for row in dict_rows:
    old_name = row["var"]
    new_name = row["new_var"]
    tipo = str(row["new_type"]).lower() if row["new_type"] else None

 # Aplicar cast de tipo
    if tipo == "byte":
        modified_df = modified_df.withColumn(new_name, F.col(new_name).cast(ByteType()))
    elif tipo == "integer":
        modified_df = modified_df.withColumn(new_name, F.col(new_name).cast(IntegerType()))
```
